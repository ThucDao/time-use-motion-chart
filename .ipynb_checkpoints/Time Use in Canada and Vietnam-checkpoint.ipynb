### Helper Functions

```python
def convert_time(minute):
    return strftime("%d-%m-%Y %H:%M:%S", gmtime(int(minute) * 60))
```

```python
def summarize_activity_data(df):
    """
    Prints summary statistics and top activities from the provided DataFrame.
    """
    print("Total records:", len(df))
    print("Unique individuals:", df['sample_id'].nunique())
    print("Time range:", df['datetime'].min(), "to", df['datetime'].max())
    print("\nTop 10 activities by frequency:")
    print(df['state'].value_counts().head(10))
```

```python
def get_safe_filepath(filepath):
    """
    Checks if a file exists. If it does, prompts the user to overwrite or rename.
    """
    while os.path.exists(filepath):
        print(f"\n[!] Warning: The file '{filepath}' already exists.")
        choice = input("Do you want to (o)verwrite it or (r)ename the new file? [o/r]: ").strip().lower()
        
        if choice == 'o':
            print(f"Overwriting '{filepath}'...")
            break # Breaks loop, returns the original filepath to overwrite
        elif choice == 'r':
            new_name = input("Enter the new file name: ").strip()
            # Ensure the new file still goes into the correct directory
            directory = os.path.dirname(filepath)
            filepath = os.path.join(directory, new_name)
            # The loop continues to check if this NEW name also exists
        else:
            print("Invalid choice. Please enter 'o' to overwrite or 'r' to rename.")
            
    return filepath
```

```python
def get_vn_activity(code):
    code = int(code)
    if code in [101, 102, 199, 201, 202, 299, 301, 302, 399, 401, 402, 499, 502]:
        return "Work"
    elif code in [198, 298, 298, 498, 598, 698, 798, 898, 998, 1098, 1198, 1298, 1398, 1498, 1598]:
        return "Travelling"
    elif code in [501]:
        return "Sell food"
    elif code in [504, 505, 506, 507, 508]:
        return "Provide services"
    elif code in [601]:
        return "Housework"
    elif code in [602]:
        return "Shopping"
    elif code in [701, 702]:
        return "Caring"
    elif code in [901, 902, 903]:
        return "Education"
    elif code in [1201, 1202, 1203, 1299]:
        return "Entertainment"
    elif code in [1301, 1302, 1399]:
        return "Sport"
    elif code == 1402:
        return "TV/Youtube"
    elif code == 1404:
        return "Surf web"
    elif code == 1501:
        return "Sleeping"
    elif code == 1502:
        return "Eating"
    elif code == 1503:
        return "Personal hygiene"
    elif code == 1506:
        return "Relaxing"
    else:
        return "Others"
```

```python
# Define the universal order for Canada and Vietnam
category_order = {
    "Sleeping": "01. Sleeping",
    "Personal hygiene": "02. Personal hygiene",
    "Eating": "03. Eating",
    "Housework": "04. Housework",
    "Caring": "05. Caring",
    "Work": "06. Work",
    "Education": "07. Education",
    "Provide services": "08. Provide services",
    "Sell food": "09. Sell food",
    "Shopping": "10. Shopping",
    "Travelling": "Travelling",
    "Sport": "11. Sport",
    "Relaxing": "12. Relaxing",
    "Entertainment": "13. Entertainment",
    "TV/Youtube": "14. TV/Youtube",
    "Surf web": "15. Surf web",
    "Others": "16. Others"
}
```

### Process and transform Vietnamese time use data

```python
print("Processing Vietnamese data...")
df_vn = df_vn.sort_values(["ID", "BEGIN"])

df_vn['BEGIN'] = df_vn['BEGIN'].apply(convert_time)
df_vn['Q401'] = df_vn['Q401'].apply(get_vn_activity)

df_vn = df_vn.rename(columns={"ID": "sample_id", "BEGIN": "datetime", "Q401": "state"})
df_vn = df_vn.dropna(subset=['sample_id', 'datetime', 'state'])
df_vn['datetime'] = pd.to_datetime(df_vn['datetime'], format='%d-%m-%Y %H:%M:%S', errors='coerce')
df_vn = df_vn.dropna(subset=['datetime'])

# Apply the category order
df_vn['state'] = df_vn['state'].map(category_order)

summarize_activity_data(df_vn)
```

**Notes:**  
Categories like âSell foodâ or âProvide servicesâ donât map directly from Canadian codes  
but can be inferred from specific self-employment or home-business entries (505).  
These can be differentiated only with further context.

```python
ca_to_vn = {
    "Work": [501, 502, 503, 504, 505, 506, 507, 599],
    "Sell food": [505],
    "Provide services": [505],
    "Housework": list(range(201, 210)) + list(range(231, 242)),
    "Eating": [150, 151, 152, 153, 154, 404],
    "Sleeping": list(range(100, 110)),
    "Personal hygiene": [126, 127, 128, 129, 130, 199, 401],
    "Shopping": [260, 261, 262, 263, 264],
    "Caring": list(range(300, 308)) + list(range(350, 354)) + [801, 802],
    "Education": [600, 601, 602, 603, 604],
    "Sport": [1001, 1002, 1003, 1004, 1005, 413],
    "TV/Youtube": [1202],
    "Surf web": [1201, 1203, 1204],
    "Entertainment": [1105, 1106],
    "Relaxing": [1301, 1302, 1303],
    "Travelling": list(range(400, 417)),
    "Others": [9999, 1304]
}
```

```python
def map_ca_activity(code):
    try:
        if pd.isna(code): return "Others"
        code = int(float(code)) # Handle float codes
        for category, codes in ca_to_vn.items():
            if code in codes:
                return category
    except (ValueError, TypeError):
        pass
    return "Others"
```

### Process and transform Canadian time use data

```python
print("Processing Canadian data...")
#ca_file = r'input\CA_time_use.csv'
#df_ca = pd.read_csv(ca_file)[["PUMFID", "STARTMIN", "TUI_01"]].dropna()

# Standardize column values
df_ca["datetime"] = df_ca["STARTMIN"].apply(convert_time)
df_ca["state"] = df_ca["TUI_01"].apply(map_ca_activity)
df_ca = df_ca.rename(columns={"PUMFID": "sample_id"}).sort_values(["sample_id", "datetime"])
df_ca = df_ca.dropna(subset=['sample_id', 'datetime', 'state'])

# Apply the category order
df_ca['state'] = df_ca['state'].map(category_order)

summarize_activity_data(df_ca)
```

### Generate a Vietnamese moving bubble chart

```python
# Create the 'output' folder if it doesn't already exist
os.makedirs("output", exist_ok=True)

# Check file existence for VN chart
vn_chart_path = get_safe_filepath("output/VN_moving_bubble_chart.html")

# Customize the speed buttons
custom_speeds = {    
    "stop": 100000, # default 100000
    "slow": 1000,   # default 1000
    "medium": 150,  # default 200
    "fast": 30,     # default 50
}

# Generate Vietnamese Moving Bubble Chart
d3_vn = D3Blocks()
d3_vn.movingbubbles(
                    df_vn,
                    filepath=vn_chart_path,
                    title='Daily Time Use in Vietnam',
                    note='',
                    figsize=(1000, 1000),
                    fontsize=20,
                    size=1,
                    speed=custom_speeds,
                    datetime='datetime',
                    sample_id='sample_id',
                    state='state',
                    center='Travelling',
                    cmap='hsv',
                    save_button=False # the SVG save button
)
```

### Create a dual chart HTML wrapper with a synchronized loading screen

```python
print("Generating dual view layout...")

# Extract JUST the filenames (without the 'output/' folder path) so the HTML iframe can load them
ca_filename = os.path.basename(ca_chart_path)
vn_filename = os.path.basename(vn_chart_path)

# Check file existence for the dual chart
dual_filepath = get_safe_filepath("output/Dual_moving_bubble_chart.html")

# Notice the 'f' before the string quotes. This allows us to inject {ca_filename} dynamically.
# But also remember to use double {{ and }} instead of { and } if it is not for inserting a variable.
html_content = f"""
<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>Dual Time Use Charts</title>
<style>
    @import url('https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap');
    
    body {{ 
        background: #f4f6f9; 
        margin: 0; 
        padding: 20px; 
        font-family: 'Roboto', sans-serif; 
        overflow-x: hidden; 
    }}

    /* --- LOADING SCREEN STYLES --- */
    #loading-overlay {{
        position: fixed;
        top: 0;
        left: 0;
        width: 100vw;
        height: 100vh;
        background-color: #ffffff;
        z-index: 9999; /* Forces the loader to sit on top of everything */
        display: flex;
        flex-direction: column;
        justify-content: center;
        align-items: center;
        transition: opacity 0.8s ease-out; /* Smooth fade out */
    }}
    
    /* A spinning circle animation */
    .spinner {{
        width: 60px;
        height: 60px;
        border: 6px solid #f3f3f3;
        border-top: 6px solid #333333;
        border-radius: 50%;
        animation: spin 1s linear infinite;
        margin-bottom: 20px;
    }}
    
    @keyframes spin {{
        0% {{ transform: rotate(0deg); }}
        100% {{ transform: rotate(360deg); }}
    }}
    
    .loading-text {{
        font-size: 24px;
        color: #333;
        font-weight: bold;
    }}
    
    .loading-subtext {{
        font-size: 16px;
        color: #666;
        margin-top: 10px;
    }}
    /* --- END OF LOADING SCREEN STYLES --- */

    /* Perfectly center the two charts */
    .container {{
        display: flex;
        flex-wrap: wrap;       
        justify-content: center;
        align-items: flex-start;
        gap: 30px;
        max-width: 1600px;
        margin: 0 auto;
    }}
    
    .chart-card {{
        background: white;
        border-radius: 12px;
        box-shadow: 0 4px 15px rgba(0,0,0,0.1);
        width: 100%;
        max-width: 800px; /* Limits how large the chart gets on giant monitors */
        display: flex;
        flex-direction: column;
        overflow: hidden;
    }}
    
    .title-bar {{
        text-align: center;
        background: #e9ecef;
        padding: 15px;
        margin: 0;
        font-size: 22px;
        font-weight: bold;
        color: #333;
        border-bottom: 1px solid #ddd;
    }}

    /* Aspect ratio perfectly hugs the chart + buttons, removing empty space */
    iframe {{
        width: 100%; 
        /* Keeps the iframe height proportional to width to fit the controls */
        aspect-ratio: 1 / 1.15; 
        height: auto;
        border: none;
        display: block;
        overflow: hidden;
    }}
</style>
</head>

<body>
    <div id="loading-overlay">
        <div class="spinner"></div>
        <div class="loading-text">Rendering two moving bubble charts...</div>
        <div class="loading-subtext">Loading 10,000+ data points for both charts. Please wait.</div>
    </div>

    <div class="container">
        <div class="chart-card">
            <h3 class="title-bar">Daily Time Use in Canada</h3>
            <iframe class="d3-frame" src="{ca_filename}"></iframe>
        </div>
        
        <div class="chart-card">
            <h3 class="title-bar">Daily Time Use in Vietnam</h3>
            <iframe class="d3-frame" src="{vn_filename}"></iframe>
        </div>
    </div>

    <script>
        document.addEventListener("DOMContentLoaded", function() {{
            // Find all the iframes we need to wait for
            const frames = document.querySelectorAll('.d3-frame');
            let loadedFrames = 0;
            const totalFrames = frames.length;
            
            // Function to check if everything is done
            function checkReadyState() {{
                loadedFrames++;
                
                // If the number of loaded frames matches the total (2 in this case)
                if (loadedFrames === totalFrames) {{
                    
                    // Add a tiny extra delay so the physics have a split second to settle
                    setTimeout(function() {{
                        const overlay = document.getElementById('loading-overlay');
                        
                        // Start the CSS fade out transition
                        overlay.style.opacity = '0';
                        
                        // Once the fade out finishes (800ms), remove it from the screen completely
                        setTimeout(function() {{
                            overlay.style.display = 'none';
                        }}, 800);
                        
                    }}, 500); 
                }}
            }}
            
            // Attach the listener to both iframes
            frames.forEach(function(frame) {{
                frame.addEventListener('load', checkReadyState);
            }});
        }});
    </script>
    
</body>
</html>
"""

# Save the dual chart HTML file
with open(dual_filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"Process completed! You can now open '{dual_filepath}' in your browser.")
```

### Post-processing: Move left panel to bottom, and hide dropdown

```python
def refine_panel_layout(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            html = f.read()

        if "viewBox" in html and "flex-direction: column-reverse" in html:
            print(f"File {filepath} has already been processed. Skipping.")
            return

        custom_code = """
        <style>
            @import url('https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap');
            
            /* Apply Roboto font on everything, including D3 SVG text */
            *, html, body {
                font-family: 'Roboto', sans-serif !important;
            }

            /* Removes internal scrollbars */
            html, body {
                margin: 0 !important; 
                padding: 0 !important;
                overflow: hidden !important; /* Removes internal scrollbar */
                background: white !important;
            }
            
            /* Make the wrapper flexible */
            #main-wrapper { 
                display: flex !important; 
                flex-direction: column-reverse !important; 
                width: 100% !important; 
                height: 100vh !important; /* Hugs the height of the iframe perfectly */
                margin: 0 auto !important; 
                align-items: center !important; 
                justify-content: flex-end !important;
                border: none !important; 
            }
            
            /* Center the bottom control panel */
            #sidebar { 
                width: 100% !important; 
                padding: 15px 0 !important; 
                border: none !important; 
                display: flex !important; 
                flex-direction: column !important; 
                align-items: center !important; 
                height: auto !important;
            }
            
            /* Allow the chart container to shrink/grow */
            #chart { 
                width: 100% !important;  
                max-width: 1000px !important; /* Match the chart figsize in Python */
                aspect-ratio: 1 / 1 !important; /* Keeps the chart perfectly square */
                height: auto !important; 
                float: none !important; 
                margin: 0 auto !important;
                display: flex !important;
                justify-content: center !important;
            }
            
            /* Clean up the UI buttons and time display */
            #current_time { 
                font-size: 40px !important; 
                color: #111 !important; 
                display: block !important; 
                opacity: 1 !important; 
                margin: 0 0 10px 0 !important; 
                font-weight: bold !important; 
                text-align: center !important;
            }
            
            #speed { 
                display: flex !important; 
                gap: 15px !important; 
                margin: 0 0 5px 0 !important; 
                float: none !important; 
                justify-content: center !important;
            }
            
            .togglebutton { 
                padding: 8px 16px !important; 
                border: 1px solid #aaa !important; 
                border-radius: 5px !important; 
                background: #fff !important; 
                color: #111 !important; 
                cursor: pointer !important; 
                font-weight: bold !important; 
            }
            
            .togglebutton.current { 
                background: #333 !important; 
                color: #fff !important; 
                border-color: #111 !important; 
            }

            /* Hide everything else to remove the leftover short line */
            #sidebar b, #ColorOptions, .clr, #day_info { display: none !important; }
        </style>
        
        <script>
            document.addEventListener("DOMContentLoaded", function() {
                
                // 1. Make the D3 SVG truly responsive using viewBox
                // We check every 100ms because D3 takes a moment to generate the chart
                var makeResponsive = setInterval(function() {
                    var svg = document.querySelector("#chart svg");
                    if (svg && !svg.hasAttribute("viewBox")) {
                        var w = svg.getAttribute("width");
                        var h = svg.getAttribute("height");
                        if (w && h) {
                            // Turn static width/height into a scaling viewBox
                            svg.setAttribute("viewBox", "0 0 " + w + " " + h);
                            svg.removeAttribute("width");
                            svg.removeAttribute("height");
                            svg.style.width = "100%";
                            svg.style.height = "auto";
                            svg.style.overflow = "visible";
                            clearInterval(makeResponsive); // Stop checking once successful
                        }
                    }
                }, 100);
                
                // 2. Fix the time format (remove Days, reset after 24 hours)
                // parseInt(m) guarantees the math works and clock display
                window.formatTime = function(m) {
                    try {
                        var mins = (parseInt(m) + 240) % 1440;
                        if (isNaN(mins)) return "04:00";
                        var hh = Math.floor(mins / 60);
                        var mm = Math.floor(mins % 60);
                        var formatted_hh = hh < 10 ? "0" + hh : hh;
                        var formatted_mm = mm < 10 ? "0" + mm : mm;
                        return formatted_hh + ":" + formatted_mm;
                    } catch(e) { return "04:00"; }
                };
                
                // 3. Keep the clock visible (bypasses internal library bugs)
                setInterval(function() {
                    var timeElement = document.getElementById('current_time');
                    if (timeElement) { 
                        timeElement.style.display = 'block'; 
                        timeElement.style.opacity = '1'; 
                    }
                }, 100);
                
                // 4. Auto-select Medium speed visually
                setTimeout(function() {
                    var mediumButton = document.querySelector('.togglebutton[data-val="medium"]');
                    if (mediumButton) mediumButton.classList.add('current');
                }, 500);
            });
        </script>
        </head>
        """

        new_html = html.replace("</head>", custom_code)
        
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(new_html)
        print(f"Success: Interface updated for {filepath}")

    except Exception as e:
        print(f"Error processing {filepath}: {e}")

# Run the processor
refine_panel_layout(ca_chart_path)
refine_panel_layout(vn_chart_path)
```
